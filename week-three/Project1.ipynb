<h1>Exploring Flight Delays for Alaska and AM WEST<h1>

```python
import pandas as pd
```

<h3>Using built-in method of Pandas to parse CSV as DataFrame</h3>

```python
df = pd.read_csv('alaska_amwest.csv')
```

```python
df
```

<h3>Adding a column called delayed_flights_percentage</h3>

```python
df['delayed_flights_percentage'] = df['delayed_flights'] / (df['ontime_flights'] + df['delayed_flights'])
```

```python
df
```

<h3>Pivot the above dataframe to easily compare delayed flights percentages between the two airlines</h3>

```python
df_pivot = df.pivot(columns='airline', index='origin', values=['ontime_flights', 'delayed_flights', 'delayed_flights_percentage'])
df_pivot
```

<h3>Finding the average of the delayed flights percentage for each airline</h3>

```python
print(df_pivot.delayed_flights_percentage['AM WEST'].mean(), df_pivot.delayed_flights_percentage['Alaska'].mean())
```
