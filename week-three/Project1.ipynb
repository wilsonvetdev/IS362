```python
import pandas as pd
```

```python
df = pd.read_csv('alaska_amwest.csv')
```

```python
df
```

<h3>Displaying all rows with delayed flights between Alaska and AM WEST</h3>

```python
delayed_flights = df[
    (df.status == 'delayed') & (df.airline == 'Alaska') | 
    (df.status == 'delayed') & (df.airline == 'AM WEST')
    ]

delayed_flights
```

<h3>Grabbing delayed flight counts for each airline</h3>

```python
alaska_delayed_flights = delayed_flights[delayed_flights.airline == 'Alaska']
alaska_delayed_flights.num_of_flights.count()
```
