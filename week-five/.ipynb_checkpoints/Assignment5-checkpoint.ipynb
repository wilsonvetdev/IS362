<h3>On February 12th, 2013, which New York area airport had the windiest weather?</h3>

<p>I want to make sure the origin column only contains EWR, JFK, and LGA by using unique().</p>

```python
weather_df.origin.unique()
```

<p>After making sure there are only three airports in the weather csv - select only the ones on Feb 12th, 2013 and sort by wind_speed.</p>

```python
feb_12_df = weather_df[(weather_df.year == 2013) & (weather_df.month == 2) & (weather_df.day == 12)]
feb_12_df.sort_values(by='wind_speed', ascending=False).head()
```

<p>After examining the above dataframe, we can conclude that LGA had the windiest weather with index <b>18417</b>.</p>
<p><a>https://en.wikipedia.org/wiki/Wind_speed</a> shows the highest speed ever recorded was 253mph in 1996. It seems unlikely for wind speed to reach 1048mph at EWR without any records on the Internet.</p>
