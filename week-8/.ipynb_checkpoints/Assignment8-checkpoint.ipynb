```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_table('auto-mpg.data', header=None)

# data.columns  = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
data['car name'] = data[1]
data = data.drop(columns = [1])
```

```python
mpg_array = []
cylinders_array = []
displacement_array = []
horsepower_array = []
weight_array = []
acceleration_array = []
model_year_array = []
origin_array = []

for i in range(len(data)):
    splitted_array = data[0][i].split(' ')
    clean_array = [item for item in splitted_array if item]
    mpg_array.append(clean_array[0])
    cylinders_array.append(clean_array[1])
    displacement_array.append(clean_array[2])
    horsepower_array.append(clean_array[3])
    weight_array.append(clean_array[4])
    acceleration_array.append(clean_array[5])
    model_year_array.append(clean_array[6])
    origin_array.append(clean_array[7])
```

```python
data['mpg'] = mpg_array
data['cylinders'] = cylinders_array
data['displacement'] = displacement_array
data['horsepower'] = horsepower_array
data['weight'] = weight_array
data['acceleration'] = acceleration_array
data['model_year'] = model_year_array
data['origin'] = origin_array
```

```python
data
```

```python
data = data.drop(columns=[0])
```

```python
data
```

```python
def convert_to_country(code):
    if code == '1':
        return 'USA'
    elif code == '2':
        return 'Asia'
    elif code == '3':
        return 'Europe'
```

```python
data['origin'] = data['origin'].apply(convert_to_country)
```

```python
data['horsepower'] = data['horsepower'].replace(['?'], np.nan)
```
