```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_table('auto-mpg.data', header=None)

data['car name'] = data[1]
data = data.drop(columns = [1])
```

```python
data
```

<h1>Auto Dataset containing vehicle data from USA, Asia, and Eruope</h1>
<p>The database had no headers. I made a new column from data[1], which contained vehicle names, and dropped the column to make it more tidy.</p>
<p>data[0] had all the other important data we needed and below is how I split up the column to make new columns.</p>

```python
mpg_array = []
cylinders_array = []
displacement_array = []
horsepower_array = []
weight_array = []
acceleration_array = []
model_year_array = []
origin_array = []

for i in range(len(data)):
    splitted_array = data[0][i].split(' ')
    clean_array = [item for item in splitted_array if item]
    mpg_array.append(clean_array[0])
    cylinders_array.append(clean_array[1])
    displacement_array.append(clean_array[2])
    horsepower_array.append(clean_array[3])
    weight_array.append(clean_array[4])
    acceleration_array.append(clean_array[5])
    model_year_array.append(clean_array[6])
    origin_array.append(clean_array[7])
```

```python
data['mpg'] = mpg_array
data['cylinders'] = cylinders_array
data['displacement'] = displacement_array
data['horsepower'] = horsepower_array
data['weight'] = weight_array
data['acceleration'] = acceleration_array
data['model_year'] = model_year_array
data['origin'] = origin_array
```

```python
data
```

```python
data = data.drop(columns=[0])
```

```python
data
```

```python
def convert_to_country(code):
    if code == '1':
        return 'USA'
    elif code == '2':
        return 'Asia'
    elif code == '3':
        return 'Europe'
```

```python
data['origin'] = data['origin'].apply(convert_to_country)
```

```python
data
```

```python
data.dtypes
```

```python
data['horsepower'] = data['horsepower'].replace(['?'], np.nan)
```

```python
data.isna().sum()
```

```python
data['horsepower'] = pd.to_numeric(data['horsepower'])
```

```python
data.dtypes
```

```python
sns.countplot(x = data['cylinders'], order = data['cylinders'].value_counts().index)

plt.show()
```

```python
data['weight'] = pd.to_numeric(data['weight'])
horsepower = data['horsepower']
weight = data['weight']

plt.plot(weight, horsepower, 'o')
plt.title('Relationship between horsepower and weight of vehicle')
plt.xlabel('weight')
plt.ylabel('horsepower')
plt.show()
```

```python
usa_data = data[data['origin'] == 'USA']
asia_data = data[data['origin'] == 'Asia']
europe_data = data[data['origin'] == 'Europe']

all_data_sum = len(data)

pie_chart_data = [len(usa_data)/all_data_sum, len(asia_data)/all_data_sum, len(europe_data)/all_data_sum]

plt.subplot(1, 2, 1)
plt.pie(pie_chart_data, labels=['USA', 'Asia', 'Europe'])

plt.subplot(1, 2, 2)
plt.bar(['USA', 'Asia', 'Europe'], pie_chart_data)
plt.subplots_adjust(wspace=0.45)
plt.show()
```
